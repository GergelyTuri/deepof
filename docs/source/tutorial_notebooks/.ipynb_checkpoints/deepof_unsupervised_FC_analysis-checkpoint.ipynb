### Load your previous project

```python
import numpy as np
import pandas as pd
import pickle

# FC data
my_deepof_project = deepof.data.load_project("./deepof_FC_project_files/")

with open("./deepof_FC_project_files/Coordinates/FC_dataset_experimental_conditions.pkl", "rb") as handle:
    exp_conditions = pickle.load(handle)
my_deepof_project._exp_conditions = exp_conditions

# Update project path to a local path
my_deepof_project._project_path = "./"
my_deepof_project._project_name = "deepof_FC_project_files"
```

```python
exp_stages = pd.read_excel("./deepof_FC_project_files/Coordinates/Fear_conditioning_stages.xlsx", index_col=0)
stage_ids = np.repeat(exp_stages["Test phase"].values, 375)
```

### Running an unsupervised analysis with default parameters

```python
# Load results
with open(
    "deepof_FC_project_files/Trained_models/train_models/deepof_unsupervised_VaDE_encoder_recurrent_encodings_input=graph_k=10_latdim=4_changepoints_False_kmeans_loss=0.0_run=0.pkl", "rb"
) as handle:
    embeddings, soft_counts, breaks = pickle.load(handle)
```

```python
embeddings = {key: val.numpy() for key, val in embeddings.items()}
soft_counts = {key: val.numpy() for key, val in soft_counts.items()}
breaks = {key: np.ones(soft_counts[key].shape[0]).astype(int) for key in soft_counts.keys()}
```

### Visualizing temporal and global embeddings

### Global separation dynamics

```python
from scipy.signal import savgol_filter
from ipywidgets import interact

@interact()
def plot_distance_between_conditions(
    # Model selection parameters
    embedding_aggregation_method=["time_on_cluster", "mean", "median"],
    hue=["Sex", "Condition"],
    sex=["both", "male", "female"],
    distance_metric=["wasserstein", "auc"],
):
    global embeddings
    global soft_counts
    global breaks
        
    # If specified, filter to keep only one sex
    if sex != "both":
        embeddings = {key:value for key, value in embeddings.items() if my_deepof_project.get_exp_conditions[key].Sex.values[0] == sex}
        soft_counts = {key:value for key, value in soft_counts.items() if my_deepof_project.get_exp_conditions[key].Sex.values[0] == sex}
        breaks = {key:value for key, value in breaks.items() if my_deepof_project.get_exp_conditions[key].Sex.values[0] == sex}

    distance_per_stage = deepof.post_hoc.condition_distance_binning(
        embeddings,
        soft_counts,
        breaks,
        {key:val[hue].values[0] for key, val in my_deepof_project.get_exp_conditions.items()},
        agg=embedding_aggregation_method,
        scan_mode="precomputed",
        metric=distance_metric,
        precomputed_bins=stage_ids,
    )

    # Concatenate both arrays and create a px compatible data frame
    stage_distance_df = pd.DataFrame(
        {
            distance_metric: distance_per_stage,
            "Time": pd.Series(stage_ids).unique(),
        }
    )
    
    fig, ax = plt.subplots(1, 1, figsize=(12, 4))
    
    # Plot the obtained distance array
    sns.scatterplot(
        data=stage_distance_df,
        x="Time",
        y=distance_metric,
        color=["#d6dbd2"],
        ax=ax,
    )

    plt.tight_layout()
    plt.show()
```

### Exploring cluster enrichment across conditions

```python
@interact
def plot_cluster_enrichment(
    # Model selection parameters
    sex=["both", "male", "female"],
    hue=["Sex", "Condition"],
    # Time selection parameters
    stage=list(pd.Series(stage_ids).unique()),
    # Visualization parameters
    normalize=False,
    save_fig=False,
):
    
    global embeddings
    global soft_counts
    global breaks    
    
        
    # If specified, filter to keep only one sex
    if sex != "both":
        embeddings = {key:value for key, value in embeddings.items() if my_deepof_project.get_exp_conditions[key].Sex.values[0] == sex}
        soft_counts = {key:value for key, value in soft_counts.items() if my_deepof_project.get_exp_conditions[key].Sex.values[0] == sex}
        breaks = {key:value for key, value in breaks.items() if my_deepof_project.get_exp_conditions[key].Sex.values[0] == sex}

    # Get cluster enrichment across conditions for the desired settings
    enrichment = deepof.post_hoc.cluster_enrichment_across_conditions(
        embeddings,
        soft_counts,
        breaks,
        {key:val[hue].values[0] for key, val in my_deepof_project.get_exp_conditions.items()},
        precomputed=stage_ids == stage,
        normalize=normalize,
    )

    fig, ax = plt.subplots(1, 1, figsize=(12, 4))
    deepof.visuals.plot_cluster_enrichment(
        my_deepof_project,
        embeddings,
        soft_counts,
        breaks,
        normalize=normalize,
        add_stats="Mann-Whitney",
        exp_condition=hue,
        verbose=False,
        precomputed=stage_ids == stage,
        ax=ax,
)
    
    fig.show()
```

### Interpreting clusters using SHAP

```python
%%time
csds_chunk_stats, hard_counts, sampled_breaks = deepof.post_hoc.annotate_time_chunks(
    deepof_project=my_deepof_project,
    soft_counts=soft_counts,
    breaks=breaks,
    #supervised_annotations=csds_OF_supervised_annotations,
    kin_derivative=1,
    window_size=13,
    include_distances=True,
    min_confidence=0.9,
    include_angles=True,
    include_areas=True,
    aggregate="mean",
    samples=1500,
)
```

```python
%%time
full_cluster_clf, cluster_gbm_performance, groups = deepof.post_hoc.train_supervised_cluster_detectors(
    csds_chunk_stats, hard_counts, sampled_breaks, n_folds=5, verbose=2,
)
```

```python
plt.figure(figsize=(8, 6))
sns.set_context("notebook")

deepof.visuals.plot_cluster_detection_performance(
    my_deepof_project,
    csds_chunk_stats,
    cluster_gbm_performance,
    hard_counts,
    groups,
    visualization="confusion_matrix",
)
```

```python
plt.figure(figsize=(8, 6))
sns.set_context("talk")

deepof.visuals.plot_cluster_detection_performance(
    my_deepof_project,
    csds_chunk_stats,
    cluster_gbm_performance,
    hard_counts,
    groups,
    visualization="balanced_accuracy",
    save=False,
)
```

```python
shap_values, explainer, data_to_explain = deepof.post_hoc.explain_clusters(
    csds_chunk_stats, hard_counts, full_cluster_clf, samples=100
)
```

```python
from matplotlib import colors as plt_colors
from ipywidgets import interact
import copy
import numpy as np

# Plot swarm plots per cluster
@interact()
def plot_shap_swarm_per_cluster(
    cluster=["all"] + list(range(10)), save=False,
):
    
    deepof.visuals.plot_shap_swarm_per_cluster(
        my_deepof_project, 
        data_to_explain, 
        shap_values, cluster, 
        save=save,
        show=False,
    )
    
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()
```
