### Cluster number selection

```python
from collections import defaultdict
import pandas as pd

# For each computed model, train a logistic regression able to predict the class (CSDS / NS) based on the vector of
# cluster population per video. Use the performance of the classifier as a metric for model selection
def get_cluster_selection_training_data(
    # Model selection parameters
    deepof_project,
    min_clusters=5,
    max_clusters=25,
    max_id=5,
):

    meta_counts = defaultdict(list)
    labels = None

    # Iteratively load all the models in question and set
    for cluster_number in range(min_clusters, max_clusters + 1):

        for run_id in range(max_id):
            
            try:
                # Load model results
                emb_path = "./NatComm_trained_models/SA_single_animal_train_models/deepof_unsupervised_VaDE_encoder_recurrent_encodings_input=graph_k={}_latdim=8_changepoints_False_kmeans_loss=0.0_run={}.pkl".format(
                    cluster_number,
                    run_id,
                )


                with open(emb_path, "rb") as handle:
                    _, soft_counts, breaks = pickle.load(handle)
                    
                soft_counts_1 = {key:val for key, val in soft_counts.items() if my_deepof_project.get_exp_conditions[key].Trial.values[0] == 1}
                breaks_1 = {key:val for key, val in breaks.items() if my_deepof_project.get_exp_conditions[key].Trial.values[0] == 1}
                soft_counts_2 = {key:val for key, val in soft_counts.items() if my_deepof_project.get_exp_conditions[key].Trial.values[0] == 2}
                breaks_2 = {key:val for key, val in breaks.items() if my_deepof_project.get_exp_conditions[key].Trial.values[0] == 2}
                
                
                # Get soft_counts
                soft_counts = {}
                
                if labels is None:
                    labels = pd.DataFrame({
                            key:deepof_project.get_exp_conditions[key].CSDS
                            for key in soft_counts_2.keys()}
                    ).T

                time_on_cluster_1 = deepof.post_hoc.get_time_on_cluster(soft_counts_1, breaks_1)
                time_on_cluster_2 = deepof.post_hoc.get_time_on_cluster(soft_counts_2, breaks_2)
                time_on_cluster_1.rename(index=trial_mapping, inplace=True)
                                
                time_on_cluster = time_on_cluster_2.sort_index() - time_on_cluster_1.sort_index()
                labels = labels.sort_index()
                                
                meta_counts["{}_{}".format(cluster_number, run_id)] = np.stack(
                    [tab for tab in time_on_cluster.values]
                )

            except FileNotFoundError:
                continue

    return meta_counts, labels.values.flatten()
```

```python
%%time
model_selection_X, model_selection_y = get_cluster_selection_training_data(my_deepof_project)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

def train_cluster_selection_classifiers(X_dict, y):

    auc_dict = defaultdict(list)

    for key, X in tqdm.tqdm(X_dict.items()):

        clf = LogisticRegression(penalty=None)        
        clf.fit(X, y)

        auc_dict[key.split("_")[0]].append(roc_auc_score(y, clf.predict_proba(X)[:, 1]))
        
    max_keys = np.max([len(v) for v in auc_dict.values()])
    for key, val in auc_dict.items():
        for j in range(max_keys - len(val)):
            auc_dict[key].append(np.nan)

    return pd.DataFrame(auc_dict).melt(var_name="number of clusters", value_name="AUC")
```

```python
%%time
import tqdm
model_selection = train_cluster_selection_classifiers(model_selection_X, model_selection_y)
agg = model_selection.groupby("number of clusters").AUC.median()
agg.index = agg.index.astype(int)
agg.sort_index(inplace=True)
```

### Paper figure layout

```python
def identify_axes(ax_dict, fontsize=48):
    """
    Helper to identify the Axes in the examples below.

    Draws the label in a large font in the center of the Axes.

    Parameters
    ----------
    ax_dict : dict[str, Axes]
        Mapping between the title / label and the Axes.
    fontsize : int, optional
        How big the label should be.
    """
    kw = dict(ha="center", va="center", fontsize=fontsize, color="darkgrey")
    for k, ax in ax_dict.items():
        ax.text(0.5, 0.5, k, transform=ax.transAxes, **kw)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create paper layout figure
sns.set_context("talk")
paper_canvas = plt.figure(layout="constrained", figsize=(24, 16)).subplot_mosaic(
    mosaic="""
           AABBCC
           AABBCC
           DDEEFF
           """,
)

# Populate panels
deepof.visuals.plot_embeddings(
    my_deepof_project, 
    embeddings_2, 
    soft_counts_2,
    breaks_2,
    aggregate_experiments=False,
    use_keys=[key for key in embeddings_2.keys()],
    samples=250,
    ax=paper_canvas["A"],
)

deepof.visuals.plot_embeddings(
    my_deepof_project,
    embeddings_1, 
    soft_counts_1,
    breaks_1,
    aggregate_experiments="time on cluster",
    exp_condition="CSDS",
    use_keys=[key for key in embeddings_1.keys()],
    show_aggregated_density=False,
    ax=paper_canvas["B"],
)

deepof.visuals.plot_embeddings(
    my_deepof_project,
    embeddings_2, 
    soft_counts_2,
    breaks_2,
    aggregate_experiments="time on cluster",
    exp_condition="CSDS",
    use_keys=[key for key in embeddings_2.keys()],
    show_aggregated_density=False,
    ax=paper_canvas["C"],
)

sns.boxplot(data=model_selection, x="number of clusters", y="AUC", color="lightgrey", ax=paper_canvas["D"])
sns.lineplot(
    data=model_selection, 
    x="number of clusters", 
    y="AUC",
    estimator=np.median, 
    ci=0, 
    linewidth=3.5,
    zorder=10,
    color="#0b7189",
    ax=paper_canvas["D"],
)
sns.scatterplot(
    x=[12], 
    y=model_selection.loc[model_selection["number of clusters"] == "7", "AUC"].median(), 
    zorder=10, 
    s=350,
    linewidth=2,
    color="#0b7189",
    ax=paper_canvas["D"],
)

deepof.visuals.plot_enrichment(
    my_deepof_project,
    embeddings_1,
    soft_counts_1,
    breaks_1,
    normalize=True,
    add_stats="Mann-Whitney",
    exp_condition="CSDS",
    exp_condition_order=["Nonstressed", "Stressed"],
    verbose=False,
    save="Trial_1",
    ax=paper_canvas["E"],
)

deepof.visuals.plot_enrichment(
    my_deepof_project,
    embeddings_2,
    soft_counts_2,
    breaks_2,
    normalize=True,
    add_stats="Mann-Whitney",
    exp_condition="CSDS",
    exp_condition_order=["Nonstressed", "Stressed"],
    verbose=True,
    save="Trial_2",
    ax=paper_canvas["F"],
)

for ax in paper_canvas:
    try:
        paper_canvas[ax].get_legend().remove()
    except AttributeError:
        continue

plt.tight_layout()
plt.savefig("CSDS_SA_figure.pdf")

plt.show()
```
